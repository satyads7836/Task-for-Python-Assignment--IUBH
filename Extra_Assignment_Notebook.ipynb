```
import glob 
import os
import pandas as pd

folder_path = os.path.expanduser('/work')
file_names = glob.glob(folder_path + "/*.csv")

li = []
for file_name in file_names:
    print(file_name)
    df = pd.read_csv(file_name, index_col=None, header=0)
    li.append(df)

# Only concatenate if li is not empty
if li:
    frame = pd.concat(li, axis=0, ignore_index=True)
    display(frame)
else:
    print("No CSV files to concatenate.")
```
